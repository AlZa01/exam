```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a dictionary to store the data for my DataFrame
data = {
  'Student_name': ['Alice', 'Chen', 'Lily', 'David', 'Tobias'],
  'Attendance': [95, 70, 85, 60, 100],
  'Mark': [90, 75, 80, 65, 95]
}
df = pd.DataFrame(data) # Creating a DataFrame for data
plt.figure(figsize=(8, 6)) # SEtting a figure size
plt.scatter(df['Attendance'], df['Mark']) # Creating a scatter plot where attendance is on the X axis and mark on the Y axis
plt.xlabel('Attendance (%)') # Labelling axes
plt.ylabel('Mark (points)')
plt.title('Correlation between Attendance and Mark') # Creating a title
plt.grid(True) # Adding a grid
plt.show() # Displaying the plot
```
